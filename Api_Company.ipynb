```python
# Import library
import requests
import json
import pandas as pd
import os
```

```python
# Save the API in the variable
url=("https://www.datos.gov.co/resource/wf53-j577.json")

# I get the response from the api
response=requests.get(url)

# Verify that if the status code is good
if response.status_code == 200:
  # print
  print(f"Good, status code is:{requests.status_codes}")

  # Save the response json
  data =response.json()

else:
  # print
  print(f"Error, status code is : {response.status_code}")

```

```python
# Convert data to DataFrame of pandas
df=pd.DataFrame(data)
df.head(3)

```

```python
# Remove unnecessary columns
df=df.drop("desc_ciiu1",axis=1)
df=df.drop("nit",axis=1)
df.head(3)
```

```python
# Cleaning dataframe
# Verify value null
print("Value null:\n", df.isnull().sum())

# Detect unique values to detect errors
print("\nValue unique:\n", df.nunique())

# Remove rows with value nulll
df_cleaned = df.dropna()

#Replacement Nan with No assignment
df_cleaned = df.fillna("No assignment")

```

```python
# Save dataframe clean

# Make path
path= "/content/dataframe/data_cleaned.csv"

# Make the folder if it does not exist
os.makedirs(os.path.dirname(path), exist_ok=True)

# Save the DataFrame in format CSV
df.to_csv(path, index=False)

# Print
print(f"File CSV dave in: {path}")

```
